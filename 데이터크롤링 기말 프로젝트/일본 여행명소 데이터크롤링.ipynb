```python
# 동적크롤링을 위한 selenium 추가 
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
# 여행정보 데이터 크롤링 
wd = webdriver.Chrome()
wd.get("https://triple.guide/trips/intro/search?tab=overseas")
```

```python
# 제이쿼리를 사용할 수 있게 추가 
wd.execute_script("var jq = document.createElement('script'); jq.src = 'https://code.jquery.com/jquery-3.6.4.min.js'; document.head.appendChild(jq);")
```

```python
# 일본 카테고리 선택 
wd.execute_script("$('.etszhc').children().eq(1).click();")
```

```python
# 일본 도시 카테고리 선택 ( 1 ~ 9 )
wd.execute_script("$('.elkmea').children().eq(1).click();")
```

```python
html = wd.page_source
soupCB1 = BeautifulSoup(html, 'html.parser')

print(soupCB1.prettify())
```
