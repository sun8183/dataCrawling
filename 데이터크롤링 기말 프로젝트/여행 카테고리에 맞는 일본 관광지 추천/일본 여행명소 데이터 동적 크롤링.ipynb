```python
# 동적크롤링을 위한 selenium 추가 ( 완성코드 )
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import time

result = pd.DataFrame()

for i in range(0, 9):
    # 여행정보 데이터 크롤링 
    wd = webdriver.Chrome()
    wd.get("https://triple.guide/trips/intro/search?tab=overseas")
    wd.execute_script("var jq = document.createElement('script'); jq.src = 'https://code.jquery.com/jquery-3.6.4.min.js'; document.head.appendChild(jq);")
    print("%d 번째 데이터 크롤링 ...... " % (i+1))
    # 제이쿼리를 읽을 텀이 필요하여 실행 
    time.sleep(2)
    # 일본 카테고리 선택 
    wd.execute_script("$('.etszhc').children().eq(1).click();")    
    # 일본 도시 카테고리 선택 ( 1 ~ 9 )
    time.sleep(2)
    wd.execute_script("$('.elkmea').children().eq("+str(i+1)+").click();")
    
    # 페이지 로딩에 필요한 시간 부여 
    time.sleep(5)
    
    html = wd.page_source
    soupCB1 = BeautifulSoup(html, 'html.parser')
    
    # 페이지 로딩에 필요한 시간 부여 
    time.sleep(5)
    
    # 카테고리의 클래스가 몇일 뒤 달라짐 .. 
    name = soupCB1.select(".iaePLH")[0].text # 변경된 클래스 : itdZDH
    cityName = [item.text for item in soupCB1.select(".iavszC")] # 변경된 클래스 : kMBrSA
    content = [item.text for item in soupCB1.select(".kqmrWM")] # 변경된 클래스 : kmXOoF
    star = [item.text for item in soupCB1.select(".hGNBCO")] # 변경된 클래스 : jbCggk
    good = [item.text for item in soupCB1.select(".cxseKY")] # 변경된 클래스 : gYqlYq
    category = [item.text for item in soupCB1.select(".kQMuew")] # 변경된 클래스 : RUqvc
    
    data = {
        '도시이름' : name,
        '관광지이름' : cityName,
        '관광지설명' : content,
        '별점' : star,
        '좋아요' : good,
        '카테고리' : category
    }
    
    df = pd.DataFrame(data)
    # 중복행 제거 
    df = df.drop_duplicates() 
    df = df.reset_index(drop=True)
    result = pd.concat([result, df])
    
print("관광지 데이터 크롤링 완료!")
```

```python
result
csv_file_path = '일본 관광지 데이터.csv'
result.to_csv(csv_file_path, index=False, encoding='UTF-8')

print("csv 파일 변환 완료")
```
